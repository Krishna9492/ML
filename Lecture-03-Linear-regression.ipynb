Machine Learning - A beginning
==============================

Before we jump into linear regression, I wanted to give a bit of an overview that might be helpful in organizing the methods we will be looking at in these notebooks.  Machine learning is about analyzing data and making predictive models that can be very usful in many applications.  Three different specific areas stand out when talking about machine learning and data analysis techniques:

1. Classification or supervised Learning: This is when we want to assign each observation or record to exactly one category or class from a predefined set of classes.  For example classify credit card transactions as "valid" or "fradulent", or classify e-mail as "spam" or "ham".  Classification is considered as an example of supervised learning, because the classes are known ahead of time, and don't need to be inferred from the data.  In classification we have a set of prelabeled records, labeled with the correct class, from which we want to fit a predictive model.

2. Clustering or unsupervised learning: In clustering we don't know ahead of time the categories or clusters that may be present in the data.  Clustering are exploratory methods, that help us find hierarchical or preexisting clusters in the data.  Since we don't provide labels, or even know the labels ahead of time, these models are considered unsupervised.

3. Recommendation: Recommendation engines, like netflix or amazon recommendations, can be seen as a type of clustering, and/or also viewed as a type of classification.  We might put them in a separate category simply because of their more recent prevelance and use in online commerce.  In a recommendation engine, we want to find movies, for example, that someone will rate highly, given their previous viewing and rating history.  So as a clustering problem, given the known ratings of a user, we would like to determine other items
that are similar to their cluster of high (and low) rated movies, to recommend (or avoid recommending).  Or we could for example sometimes state this as a classification problem, where users have rated movies as 1, 2, 3 or 4 stars (for example) and we want a model that will categorize unseen movies into their most likely star rating, given a particular users data.

In most machine learning problems, we begin with a data set contining multiple elements, records or instances.  Each instance consists of several (sometimes many thousands) of attributes, also called features.  For classification or prediction tasks, usually one of the features or attributes is the thing we are interested in predicting or categorizing.  In such a case, the other attributes or features may be considered as inputs (or independent variables) for any predictive model we create, while the thing we are trying to classify is the output, class or dependent variable if we are in a scientific context.

When creating a model, for complex data, we often need to get a feel for how well the model will perform with unknown data.  Sometimes the data is real hard to model in order to classify, and we may be very happy getting 60% correct.  In other contexts, this might be abysmal. When using supervised learning to create a model, we often split our data into two sets, a training set and a test set.  We create the model with the training set.  But then instead of seeing how well the model performs on the data it was trained with, we instead see how well it performs on the test set, a set of data it has not seen, and was not trained with.  The test set does actually have the preexisting category labels, so we can them measure how well our predictive model does on data it has not seen before.  This technique is known as cross validation, and is very common.

Cross validation is often done for severl reasons.  Unfortunately, it can often be too easy to overfit a model to one particular set of data.  An overfitted model will thus perform spectacularly well on the data it was trained with, but will perform very poorly on unseen data.  Cross validation helps us determine if overfitting has occurred.

Linear Regression
=================

In the abstract, regression is a very simple concept: you want to predict one set of numbers given another set of numbers.  For example, acturies want to predict how long a person will live given their smoking habits (and other factors, very important for the insurance industry), or we might want to predict a persons weight given their height.  Thus regression is a type of supervised learning: we have a known set of factors, with the correct outputs for them.  Our task is to create a model so that if we se an unknown example of the data, we can measure the factors or predictors and make an estimate or prediction of the variable of interest.  In regression type work, the inputs are often called factors, features or predictors, and the outputs are the predictions.  This is also basically the same concept in scientific research, where the inputs are the independent measures, and the outputs are the dependent measures.

Because we are predicting numbers, we want to make strong statements about the relationship between the inputs and the outputs.  We might want to say, for example, when the number of packs of cigarettes a person smokes per day doubles, their predicted life span gets cut in half.

Accuracy of Prediction
----------------------

Before we begin building and comparing regression prediction models, we need a way of defining how good are the predictions of the models we are making.  There are a lot of reasonable ways to define the accuracy of prediction, but the most common and dominant method in machine learnining is to the the squared error.

If you are trying to predict a value $y$ (the true output) and you guess $h$ (your hypothesis about y), the the squared error of your guess is simply $(y - h)^2$.  If we have many such predictions, we can sum them up to get what is known as the sum squared error (SSE).  There are many reasons why the squared error is reasonable for estimating how good our set of predictions are.  For one, we have to square the values, or else some predictions would be too high, and some too low, and when we sum them together many would cancel each other out and we would end up with a value of 0.  (Instead of squaring the difference, you will sometimes see people take the absolute value of the difference, to correct this problem).

Lets calculate the sum of the squared error using the mean value of age at death as our model.

```python
def sum_squared_error(output, hypothesis):
    """output and hypothesis are expected to be equal sized NumPy or NumPy-like vectors.
    We calculate the difference between each output and hypothesis, square the result and sum
    together to return as the result.
    """
    return np.sum( (output - hypothesis)**2.0 )
```

```python
# create an array same size as death_age where all values are the mean age at death as the hypothesis
hypothesis = np.empty(death_age.shape)
hypothesis.fill(death_age.mean())
sse = sum_squared_error(death_age, hypothesis)
print "sum squared error:", sse
```

Another common measure you might see used insted of summing all of the squared errors, is to take the mean of the squared error (MSE).  This method can reduce the effect that outliers have on the estimate of the total error, which may or may not be a good thing.  In any case calculating the mean squared error is also very easy with python.

```python
def mean_squared_error(output, hypothesis):
    """output and hypothese are expected to be equal sized NumPy vectors or list-like structures.
    We calculate the difference between each output and hypotheses, square this, and find
    the average (mean) of the squared differences to return as the result.
    """
    return np.mean( (output - hypothesis)**2.0 )
```

```python
mse = mean_squared_error(death_age, hypothesis)
print "mean squared error:", mse
```

Optimizing Our Hypothesis
-------------------------

We now have a measure of the performance (or the error if you like) of how our model does at our prediction task.  So far we have only looked at the simplest of models, predicting the average age of death for any new data.  This in and of itself may not be enough to convience you that the mean value of 72.72 is the best we can do (without any other information) in predicting age.  

To give us an intuivite feel that this might be the case, we could calculate the MSE on our data set for a range of "model" values, and see where we reach a minimum value in the error.  This is the first example of a type of optimization approach, we will explore a parameter and try and find where it is at its minimum value.  We will look at optimization techniques in a later lecture.  But for now, we see the basics of the approach, where we visualize the performance of a model in some parameter space, and we want to find the location in that space where the performance is minimized (when minimizing error in prediction, for example), or sometimes where it is maximized (getting the highest correlation score, for example).

We can easily calculate the MSE for possible guesses ranging from, say, 60 to 85, to see what happens.

```python
guesses = np.linspace(60.0, 85.0, num=5000)
errors = [mean_squared_error( death_age, np.ones(death_age.shape) * g ) for g in guesses]
```

```python
plt.figure(figsize=(8,8))
plt.plot(guesses, errors, 'b-')
plt.xlabel('Age (model hypothesis)')
plt.ylabel('MSE')
```

As you can see from looking at this figure, the minimum does occur at the mean age, indicating that the best possible hypothesis we can make (using no other information) is the mean age of death of the original data.

One important implication of this: any other model using extra information will need to improve on this performance in order for us to consider that the model has any extra predictive value.

Regression Using Dummy Variables
--------------------------------

So how can we use the extra smoker/nonsmoker information to improve our model.  One simple idea is to break our hypothesis/prediction into two parts.  We will use the mean age of death of smokers, if we are making a prediction for a smoker, and use the mean age of death of nonsmokers if we are looking at a nonsmoker.  We have already broken the data into these two groups based on the Smokers dummy variable, so we can see the mean age of death of each group:

```python
print "Mean age of death, smokers:", death_age_smokers.mean()
print "Mean age of death, nonsmokers:", death_age_nonsmokers.mean()
```

Lets make an array that implements this hypothese and calculate its MSE.  The function will take the smokes dummy variable array, and return a hypothesis where all smokers have a gues of 70.192 years, and all nonsmokers have a guess of 75.254 years.  The example will use some NumPy vector operations to perform this set of operations.

```python
hypothesis = np.empty(death_age.shape)
hypothesis[where(smokes==0)] = death_age_nonsmokers.mean()  # fill all indexes of nonsmokers with the mean age of death of nonsmokers
hypothesis[where(smokes==1)] = death_age_smokers.mean()     # likewise fill all indexes of smokers with their mean age
mse_v2 = mean_squared_error(death_age, hypothesis)
print "MSE without extra information: ", mse
print "MSE taking smoking/nonsmoking into account: ", mse_v2
```

As you can see by looking at the MSE, our predictions really do get better after we include more information.  

Probably the most popular measure of error in the ML literature is not the MSE, but the root mean squared error (RMSE).  This can be calculated simply by taking the square root of the MSE.  The RMSE, by taking the square root, returns the value back to its original units of measure.  So for example, the RMSE of our two models are:

```python
print "RMSE without extra information: ", mse**0.5
print "RMSE taking smoking/nonsmoking into account: ", mse_v1**0.5
```

The RMSE is helpful because we can interpret it more directly.  For example, the average error we achieve for our age of death without any extra information is about 5.73 years (+ or - from the true age), but the average error for our model with the smoker/nonsmoker information has improved to +/-5.14 years.  This represents about a 10% improvement in our predictions.

Linear Regression in a Nutshell
-------------------------------

So now we have a mechanism for incorporating dummy variables.  We saw an example of a binary variable, but we could do the same thing on any category variable that has a finite discrete set of values, and see if using that information improves our predictions.  But how can we use richer information in our predictions.  By richer, we mean continuous values such as heights and weights.  Also, we would like to be able to use more than 1 source of information simultaneously in order to build our  prediction model.

Using all of the information we have may not be an easy task.  One thing we can do is to see which dimensions (features) of our information matter most, and only use the most important factors (see the lecture on Principle Component Analysis PCA).  Another thing that we may need to do is to make simplifying assumptions in order to get things to work.  These assumptions underlie linear regression.

The two biggest assumptions we will make when using linear regression to predict outputs are:

1. Separability/addivity:  We assume each factor is independent of other factors.  This is often the case, and even when it isn't it still might not affect our hypothesis too much to make this assumption.

2. Linearity: We assume that the data is linear or monotonic, meaning if one of the inputs go up it always makes the output go up (or down if negatively correlated).  We can never have the case where increasing the input causes the output to go up for a bit, then down some, then up some (this is nonmonotonic).

Keeping these assumptions in mind, lets start working through a simple example of linear regression.  Let's go back to using our heights and weight data set to show how linear regression works when we have complex input features.

```python
heights_weights = pd.read_csv('data/01_heights_weights_genders.csv')
heights = heights_weights.Height
weights = heights_weights.Weight
heights_weights.describe()
```

We will be using height as the input value or feature, and weight as the output or value to be predicted.  If we fit a line
to the data we will see the line that a simple linear regression model will use for this 2-dimensional data. 

```python
plt.figure(figsize = (8, 8))
plt.plot(heights, weights, '.', mfc = 'white', alpha = .2)
plt.xlabel('Height (in)')
plt.ylabel('Weight (lbs)')
# use numpy least square polynomial fitting to fit a 1D line to data
p = np.polyfit(heights, weights, 1)
f = np.poly1d(p)
w = np.linspace(heights.min(), heights.max()) 
predictions = f(w)
plt.plot(w, predictions, '-k')
```

Looking at this plot should convince you that using the best fit line to predict a person's weight given their height should work pretty well.

Lets use the SciKitLearn module's LinearRegression objects to create a model of our height by weight data.  The SciKitLearn LinearRegression object fits your data as above, but allows you to get much more information about the linear regression models you create.  First we will create our linear regression objects, then fit our data to a model.  Then we will show how, using the created object, to get various information, make our predictions, etc.

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
#heights = heights.reshape((heights.shape[0],1)) 
regr.fit(heights.reshape((heights.shape[0],1)), weights) # regr objects needs an array, not a vector, so reshape heights to make it happy
```

First of all, we can get the coefficients associated with this linear fit.

```python
print "Slope = ", regr.coef_
print "Intercept = ", regr.intercept_
```

These values should be familiar to you from simple geometry.  We can interpret the slope here as meaning that for every 1 inch increase in a persons height, we expect to see about a 7.7 lb increase in their weight.  The intercept is a little tricker, it says that a person who is 0 inches tall would weight -350 lbs.  This does point out some of the limitations of such models, we need to know how they were created.  In this case, we only created the model from the height and weight data of normal healthy adults, thus we should not expect it to work for children, or for populations that fall outside of the normal variation of adult heights and weights.

From the linear regression object, we can easily get our predictions for all of our data.  We can also use the predictions to calculate all of the errors (the differences between prediction and actual output).  This list of errors is known as the residuals:

```python
predictions = regr.predict(heights.reshape((heights.shape[0],1)))
errors = weights - predictions
```

These errors are called residuals because they are the part of our data that's left over after accounting for the part that a line can explain.  A common way to diagnose any obvious mistakes we're making in linear regression is to plot the residuals against the truth:

```python
plt.figure(figsize=(8,8))
plt.plot(weights, errors, '.', mfc = 'white', alpha = .2)
plt.xlabel('Weight (lbs)')
plt.ylabel('residual error')
```

There appears to be no obvious structure in the residuals (a nice round blob, thought is it slightly positively correlated?), which is what we want.  If we see some systematic correlation of the residuals, this is usually a bad thing.  A model divides the word into the signal, and the noise which doesn't help with the prediction.  If the residuals seem to have some signal in them, then your model isn't powerful enough to extract all of the signal in the data and leave behind only real noise.  

Here is an example where a linear model isn't appropriate, so you can see clearly when plotting the residuals is showing something wrong with the model:

```python
x = np.linspace(1, 10, 100)
y = x*2.0

bad_regr = linear_model.LinearRegression()
bad_regr.fit(x.reshape( (x.shape[0], 1) ), y)

predictions = regr.predict(x.reshape((x.shape[0],1)))
errors = y - predictions

plt.figure(figsize=(8,8))
plt.plot(y, errors, '.', mfc = 'white', alpha = .9)
plt.xlabel('$y = x^2$')
plt.ylabel('residual error')
```

Having the residuals is great, but we would like to have a single number that will tell us how we are doing with our model.  The solution to this we have already seen, using the squared error, and summing or taking the mean of them, and finally taking the square root of the MSE if we want to have a measure in the same units as the original output.  The residuals are actually the first step towards calculating a SSE, MSE or RMSE measure, since they are the differences between the prediction and the actual value.  We can calculate all three of these by hand from the residuals:

```python
predictions = regr.predict(heights.reshape((heights.shape[0],1)))
errors = weights - predictions
squared_errors = errors**2.0

sse = np.sum(squared_errors)
mse = np.mean(squared_errors)
rmse = mse**0.5

print "SSE = %f,  MSE = %f,  RMSE = %f\n" % (sse, mse, rmse)
```

The simple sum of squared errors (SSE) can be problematic.  As the size of the data being modeld increases, the SSE will increase.  So it can't be used for comparing performance of models on different data, or even data with different numbers of data points in them.  The mean squared error (MSE) solves this, making it easier to use to compare different models.  And as we've mentioned, the RMSE then makes the number obtained interpretable in the same units that we are predicting.  For our linear heights to weights regression, we see that our model is averaging about 12.2 lbs of error in its predictions.

One problem with RMSE is it is still hard to tell whether the error represents good or poor performance on the data.  When running linear regression, the solution to this problem is to find the $R^2$ (pronounced "R squared") measure.  The idea of $R^2$ is to see how much better your model does than we'd do if we were just using the mean to predict (remember from the beginning of this notebook), at the low end or a score of $R^2$ of 0.0.  And comparing it to how well we are doing compared to perfect prediction, a $R^2$ of 1.0.  $R^2$ will always be between 0 and 1,  If you are doing no better (or worse) than the mean performance, your $R^2$ score will be 0.  And if you predict every point perfectly, your $R^2$ will be 1.  

We can easily compute this by hand, we just need the SSE of our model, and compare this to the SSE of using the simple mean of the data as a prediction model on the same data.  Lets see our $R^2$ score for our heights to weights linear regression model.

```python
linear_regression_sse = sse # from previous cell
mean_sse = np.sum( (weights - np.ones(weights.shape) * weights.mean())**2.0 )
r2 = 1.0 - (linear_regression_sse / mean_sse)
print linear_regression_sse, mean_sse
print "The R^2 score we calculated was: ", r2
```

We can get this same $R^2$ score result much easier using our linear regression object that has already been fitted to our data:

```python
print 'R^2 score of model:', regr.score(heights.reshape((heights.shape[0],1)), weights)
```

Linear Regression with More Factors
-----------------------------------

So far we have shown a linear regression using only 1 factor.  But the linear regression objects can be used with many factors to try and predict a continuous output value.  If we have two factors, the model does not form a line, but instead a plane in the 3 dimensional space (2 input factors by the 1 output prediction we are making).  And it works for higher numbers of factors, though above 3 dimensions we can't visualize the hyperplane of the linear regression solution thus generated.   Here is an example of another data set with 2 input factors, where the linear regression model will create a minimal hyperplane model of the 2 input factors.  

We will work with the Parkinsons Telemonitorin" dataset of the University of Oxford. The original study used a range of linear regression methods to predict the clinician's Parkinson's disease symptom score on the UPDRS scale
We load the dataset "Parkinsons Telemonitoring" using the numpy loadtxt function.
The columns are separated by ' ' delimiter, which we pass to the loadtxt function.

```python
parkinsons_data = np.loadtxt("data/artifical_lin.txt")
parkinsons_data.shape
```

There are 500 data points (rows), each with 3 factors (columns).  The factors in the first 2 columns will be the factors, or inputs, we will use to fit the model, and we will be trying to predict the data in the 3rd column, the outputs.

```python
factors = parkinsons_data[:, :-1]
outputs = parkinsons_data[:, -1]
print factors[:10, :]
print factors.shape
print outputs[:10]
print outputs.shape
```

Before doing the linear regression, we will show another technique that is typical when performing a machine learning analysis to create a model.  We will split up the data into training sets and testing sets, and measure the performance of our modle on how well creating it only from the training set subportion of the data works to model the other data we split off into the test set.  We need to be careful if the data is ordered, to shuffle up the data, so that we create training and testing sets that are representative of the whole range of data in the inputs.  This technique is known as cross validation, and is used very commonly when developing ML models.

We first start by shuffling up the data randomly.  The random_state parameter acts like a seed for a random number generator here.  If we shuffled the factors and outputs again using same random_state, we would obtain the same shuffling.  Note that the ordering of data item rows is kept for the shuffle.  For example the row 0 of the factors input corresponds to the resulting row 0 in the outputs.  After shuffling if row 0 was moved to row 30, it will be move to row 30 in both the factors and outputs, in order to maintain the correspondence.

```python
from sklearn.utils import shuffle
factors, outputs = shuffle(factors, outputs, random_state=1)
print factors.shape
print outputs.shape
```

Now we split the data into training and testing sets to perform the cross validation.  This is easy to do in Python using array slices.

```python
# first split the inputs
train_set_size = factors.shape[0] / 2
print train_set_size
factors_train = factors[:train_set_size, :]  
factors_test = factors[train_set_size:, :]   
print(factors_train.shape)
print(factors_test.shape)

# now split the outputs
outputs_train = outputs[:train_set_size]   
outputs_test = outputs[train_set_size:]    
print(outputs_train.shape)
print(outputs_test.shape)
```

Let's look at the data in a 3d plot. There appears to be some linear relationship in the data (which is good if we hope for our linear regression model to work):

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(12,12))
ax = Axes3D(fig)
ax.scatter3D(factors_train[:500, 0], factors_train[:500, 1], outputs_train[:500])
ax.view_init(6,-20)
plt.show()
```

Create the linear regression model and fit the model using the training set.

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
regr.fit(factors_train, outputs_train)
```

Display the coefficients as we did before.  Notice we now have two coefficients, or slopes, as this linear model corresponds to a 2D plane, and the two coefficients give the slopes in the two dimensions of our corresponding input factors.

```python
print regr.coef_
print regr.intercept_
```

Now lets calculate the sum square, mean square and root mean square errors of the model against both the training set used to create the model, as well as the test set.  By looking at the performance on the test set, we will get an idea of how this model will perform for future unknown data.

```python
train_sse = np.sum( (regr.predict(factors_train) - outputs_train) ** 2.0 )
train_mse = np.mean( (regr.predict(factors_train) - outputs_train) ** 2.0 )
train_rmse = train_mse**0.5

print "Model results against the training set used to create model SSE=%f  MSE=%f  RMSE=%f" % (train_sse, train_mse, train_rmse)

test_sse = np.sum( (regr.predict(factors_test) - outputs_test) ** 2.0 )
test_mse = np.mean( (regr.predict(factors_test) - outputs_test) ** 2.0 )
test_rmse = test_mse**0.5

print "Model results against the test set SSE=%f  MSE=%f  RMSE=%f" % (test_sse, test_mse, test_rmse)
```

The results on the test set are, of course, not as good as on the set the model was trained with, but still look pretty good.  Lets look at the $R^2$ value to get a feel for how well this linear regression does relative to a simple mean model.

```python
print "R^2 of training set:", regr.score(factors_train, outputs_train)
print "R^2 of test set:", regr.score(factors_test, outputs_test)
```

Excellent.  We didn't mention before, but often $R^2$ is used informally as an interpretation of the percent of signal the model is able to use for prediction.  Here we are getting 82% of the signal or so using linear regression.

Finally, lets visualize the resulting linear regression plane, as we did before with the regression line for weight.  Here we will see the model plane generated for this data by the linear regression. 

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(12,12))
ax = Axes3D(fig)


ax.scatter3D(factors[:500, 0], factors[:500, 1], outputs[:500])

# here we create plane which we want to plot, using the train data and predictions (you don't need to understand it)
range_x = np.linspace(factors_train[:, 0].min(), factors_train[:, 0].max(), num=10)
range_y = np.linspace(factors_train[:, 1].min(), factors_train[:, 1].max(), num=10)
xx, yy = np.meshgrid(range_x, range_y)
zz = np.vstack([xx.ravel(), yy.ravel()]).T
pred = regr.predict(zz)
pred = pred.reshape(10, 10)

ax.plot_surface(xx, yy, pred, alpha=.1)  # plots the plane
ax.view_init(6,-20)
plt.show()
```
