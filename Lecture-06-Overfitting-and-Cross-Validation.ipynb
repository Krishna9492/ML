Methods for Preventing Overfitting
==================================

We have several times mentioned the problem of overfitting and shown brief examples of using cross-validation to try and address the issue.  In this lecture we will go into the method of cross-validation in a bit more detail, and discuss more the problem of overfitting a model.

Before we can prevent overfitting, we ought to have some specific meaning and understanding of what we mean by the term of overfitting.  Until now we have mostly relied on your intuitive sense that overfitting happens when the model we have created matches part of the noise in the data set, rather than the true underlying signal.  But if we don't know what the truth is, how can we tell when this has occured, that we are overfitting and modeling some of the random noise?

The trick is to make clear what we mean by truth.  For our purposes, a predictive model is close to the truth when its predictions about future data are accurate.  Or in other words, we will formalize the idea of overfitting by basically the method we use in cross-validation, examining how the model performs on unseen and untrained data. Of course we don't have access to future data right now, and we may not want to wait for future data to be gathered before we make a determination of whether our model has a good fit, or appears to be overfitted.  This implies that we instead use the data we have, and only train with part of it, then test with the other, untrained on part of the data set.  This splitting up of the data we have, to measure expected performance on future unseen data, is known as cross-validation.

Arguably, cross-validation is simply an instantiation of the scientific method that we were all taught as children:

1. formulate a hypothesis,
2. gather data, and
3. test it.

There's just a bit of sleight of hand because we don't formulate a hypothesis based on existing data and then go out and gather more data.  Instead, we ignore part of our data while formulating our hypotheses, so that we can magically rediscover that missing data when it comes time to test our predictions.

Lets show cross-validation once again on a toy problem with data we generate.  We will create a signal using a simple sine wave, but add a bit of random gaussian noise to the signal.

```python
x = np.arange(0, 1, 0.01)
y = sin(2 * pi * x) + np.random.normal(0, 0.1, size=x.shape)

plt.figure(figsize=(10,8))
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('sin(x) + noise($\sigma = 0.1$)')
plt.title('sin() signal with noise')
```

Our goal is to build a model of the function, that predicts the actual signal, and doesn't overfit and model some of the noise.  We will use regression modeling again, but we will build models with polynomial functions.  A polynomial function of degree 1 is simply a linear function, thus a linear regression.  But we can fit polynomials with larger degrees.  We will fit models from degree 1 to 12 and assess our performance.

Lets first split our data into two parts, a training set and a test set.  We will test performance on both the trained set, and the test set, to get a feel for how overfitting works and what it looks like.  For this example we will split the data exactly in half, 50% of the data in the training set and 50% in the test set.  For some applications its better to use more data for the training and less for the test set, because the more data we have the better the prediction.  How to best to split depends very much on the problem and the data, and often has to be experimented with to find the right balance.  The extreme end of cross-validation is where we take all of the data but one point, and then see how well the model predicts the excluded point.  This is known as leave-one-out cross-validation.  Often we perform leave-one-out many times, removing one point and testing the prediction repeatedly, and averaging the result.

We will first once again make use of the `shuffle()` function from `sklearn.utils`.  It is usually important to randomly split the data
between the test and training set, since different parts of the input/output space might have different characteristics.  `shuffle()` here simply randomly rearranges all of the items in the arrays it is passed, but if you pass in more than 1 array, the order is preserved, so that the `y` corresponding to each $f(x)$ still correspond correctly.  For example:

```python
from sklearn.utils import shuffle

print x[:5]

# what happens when we shuffle the ordering, for example the x values are no longer ordered
x, y = shuffle(x, y, random_state=1)
print x[:5]

plt.plot(x, y, '.')
```

Notice that before the first five `x` values are in the original order that we created them in using the `arange` function, but after they have been randomized.  I plot the data again, to verify that the ordering of corresponding `x,y` values is still intact.  If the ordering was not kept in tact by `shuffle`, you would no longer see the $sin(x)$ function relation between `x` and `y`.

As we noted before, splitting into train and test sets once we successfully randomize is trivial in a high-level scripting language like Python that supports array slicing:

```python
train_set_size = x.size / 2

x_train = x[:train_set_size]
y_train = y[:train_set_size]
print x_train.shape, y_train.shape

x_test = x[train_set_size:]
y_test = y[train_set_size:]
print x_test.shape, y_test.shape
```

As you can see, our sin wave data has now been split in half, into train and test sets.

It turns out, that in addition to the handy `shuffle()` function, SciKit learn has a utility function to perform the whole cross-validation split for you.  So we can simply do this (NOTE that the function is performing both the split, as well as the random shuffling):

```python
x = np.arange(0, 1, 0.01)
y = sin(2 * pi * x) + np.random.normal(0, 0.1, size=x.shape)

from sklearn import cross_validation
x_train, x_test, y_train, y_test = cross_validation.train_test_split(x, y, test_size=0.5, random_state=1)
print x_train.shape, y_train.shape
print x_test.shape, y_test.shape
print x_train[:5], x_test[:5]
```

Here we will perform a regression analysis using the polyfit method from NumPy which performs least squared fitting to a polynomial of a given degree.  For example, to fit polynomial of degrees 1, 2 and 4 to our data, we would do polyfits like this:

```python
plt.figure(figsize=(12,8))
plt.plot(x_train, y_train, 'bo', fillstyle='none')

# degree 1 (linear fit)
p = np.polyfit(x_train, y_train, deg=1)
print p # the coefficients of the fit, these can be passed to np.polyval to create a prediciton/model
x_pred = np.arange(0, 1, 0.01)
y_pred = np.polyval(p, x_pred) # 
plt.plot(x_pred, y_pred, 'r-', label='degree = 1 (linear model)')

# degree 2 (quadratic fit)
p = np.polyfit(x_train, y_train, deg=2)
print p # the coefficients of the fit, these can be passed to np.polyval to create a prediciton/model
y_pred = np.polyval(p, x_pred) # 
plt.plot(x_pred, y_pred, 'g-', label='degree = 2 (quadratic)')

# degree 4 
p = np.polyfit(x_train, y_train, deg=4)
print p # the coefficients of the fit, these can be passed to np.polyval to create a prediciton/model
y_pred = np.polyval(p, x_pred) # 
plt.plot(x_pred, y_pred, 'c-', label='degree = 4')

# degree 12 
p = np.polyfit(x_train, y_train, deg=12)
print p # the coefficients of the fit, these can be passed to np.polyval to create a prediciton/model
y_pred = np.polyval(p, x_pred) # 
plt.plot(x_pred, y_pred, 'm-', label='degree = 12')

plt.xlabel('x')
plt.ylabel('sin(x) + noise')
plt.grid()
plt.legend(loc=3)
```

You should note two things from the previous figure.  Higher degrees do fit better, to a point.  But very high degrees can be overfit, and follow curves or paths that are not actually there in the data (they are artifacts of the noise).

Lets perform a little experiment/analysis to determine the `best` fit of a polynomial to this data.  We will examine all degrees from 1 to 12, and test performance on both the training set, as well as against the untrained test set.  We will use the RMSE to measure how well a particular model does against the data.

```python
def rmse(y, h):
    """calculate the Mean Squared error and return the square root, to put back into original units of measure
    y is the correct data, and h is the hypothesis or predictions made about the data.
    """
    return np.mean((y - h)**2.0)**0.5
```

```python
# save results in perf data frame for analysis
degree = 12
perf = pd.DataFrame(columns=['train', 'test'], index=range(1,degree+1))

# perform experiments on degrees from 1 to 12, get performance on both the original training data, and our test data
for d in range(1, degree+1):
    p = np.polyfit(x_train, y_train, deg=d)
    h = np.polyval(p, x_train)
    train_perf = rmse(y_train, h)
    h = np.polyval(p, x_test)
    test_perf = rmse(y_test, h)
    perf.ix[d, 'train'] = train_perf
    perf.ix[d, 'test'] = test_perf
```

Let analyze our results.  First lets plot how RMSE varies as a function of d for the train and test data:

```python
plt.figure(figsize=(12,8))

plt.plot(perf.index, perf.train, 'b-', label='training set')
plt.plot(perf.index, perf.test, 'r-', label='test set')
plt.grid()
plt.legend()
plt.xlabel('Degree')
plt.ylabel('RMSE')
```

You can get different results depending on the random noise generated, so try regenerating a couple of times.  But in general you should usually see a couple of things:

1. The performace on the test set should usually be worse than the training set, as we would expect since the model was trained on the training data.  This isn't always the case, the performance on the test data can occassionally (by chance) outperform performance on the original trained data, but not often.
2. Usually around degree 3 or 4 you see a big improvement in the RMSE predictions.
3. The performance on the training data, after the initial big improvement, will usually slowly continue to improve (decrease).
4. Usually sometime after degree 9 or 10, you will see that the test set performance begins to deteriorate rapidly.  In fact, after degree 12 the performance will plumeet.  If you try running the experiments for degrees higher than 15 or so, you will easily see this.

The rapid deterioration of performance is due to overfitting of the polynomial to the noise in the data.  The measurement of the performance as a function of the test data, and the detection of performance becoming significantly worse on the test data, is the definition of what we mean by overfitting.

Also one other thing to note in the previous example.  This is the second example (I believe) we've seen of a type of optimization procedure.  Here we trying to optimize the degree d parameter of the polynomial that provides the best fit to our signal.  We will discuss this more in the Optimization lectures, but this is another good example of the general technique.  We view the parameter space we are trying to optimize in terms of a performance (or fitness) measure, and try to find the parameter in the space where the performance is maximized (or in this case, where the error in performance is minimized).

Overfitting and Best Parameters for KNN Classification
------------------------------------------------------

Here we will show a more complex example of using cross-validation for the KNN classifier that you worked with in your second assignment.  There we introduced 2 parameters in trying to build a KNN classifier, the number of clusters the KNN classifier should form, and the type of distance measure that the classifier should use.  Lets perform a similiar experiment as we just did on this more realistic example.

I want to perform the analysis/experiments on the Iris data set again, but that dataset is a bit small.  Therefore I am going to use another variation of a technique, oversampling.  We will actually create a larger dataset from the Iris database.  I am going to randomly draw samples from the original data, with the possibility of drawing a data point many times to the new sample.  I am also going to add some noise to each samle that I draw.  This is not a normal technique, but it will give use a larger dataset, that is a semi-noisy variation of the original data set.

```python
# create a new oversampled data set of the Iris data.  I want to end up with 9999 new data items, evenly distributed over
# the 3 classes, with a moderate bit of noise added to each item.
from sklearn import datasets
iris_orig = datasets.load_iris()

iris_data = np.zeros((9999, 4))
iris_target = np.zeros(9999)

ix = 0
for type in range(0,3):
    for n in range(0, 3333):
        # this selects index of one of the original items of 'class' type (0, 1 or 2) at random
        random_item_ix = np.random.choice(where(iris_orig.target == type)[0])
        
        # and this statement adds gaussian noise with std=1.0 to the 4 input/features we randomly selected
        iris_data[ix] = iris_orig.data[random_item_ix] + np.random.normal(loc=0.0, scale=1.0, size=(4,))
        iris_target[ix] = iris_orig.target[random_item_ix]
        ix += 1
        
print iris_data.shape
print iris_target.shape

print iris_data[:5,:]
print iris_target[:5]
```

```python
# now we randomize and split into training and test sets.  I'll use 2/3 of data for training and 1/3 for test.
# notice where how important it is to randomize the ordering.  All of the data before this are grouped, the first 3333
# items are of class 0, the next 3333 of class 1, etc.  If we did the split without shuffling this data, we would end up
# with all of the 0 and 1 class in the train set, and only the 2 class in the test 
iris_data_train, iris_data_test, iris_target_train, iris_target_test =  cross_validation.train_test_split(iris_data, iris_target, test_size=0.3333)
print iris_data_train.shape, iris_data_test.shape
print iris_target_train.shape, iris_target_test.shape
```

```python
# an example of what we will be testing systematically.  
# We define our fitness function, in this case simply counting the number of misclassified items
def num_misclassified(target, pred):
    """Simply count the number of predictions that are not correct (not == to the target).  We assume
    that target and pred are numpy array types and return a single integer count of the misclassified items
    """
    return (target != pred).sum()

from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3, weights='distance')
knn.fit(iris_data_train, iris_target_train)

pred = knn.predict(iris_data_train)
train_misclassified = num_misclassified(iris_target_train, pred)
print "training items misclassified: ", train_misclassified

pred = knn.predict(iris_data_test)
test_misclassified = num_misclassified(iris_target_test, pred)
print "test items misclassified: ", test_misclassified
```

```python
# now we'll test number of neighbors from 2 to 50, each value using 'uniform' or 'distance' weights for calculating neighbors
from sklearn.neighbors import KNeighborsClassifier
perf = pd.DataFrame(columns=['n_neighbors', 'weights', 'train', 'test'])

for n_neighbors in range(2, 51):
    for weights in ['uniform', 'distance']:
        knn = KNeighborsClassifier(n_neighbors=n_neighbors, weights=weights)
        knn.fit(iris_data_train, iris_target_train)

        pred = knn.predict(iris_data_train)
        train_misclassified = num_misclassified(iris_target_train, pred)
    
        pred = knn.predict(iris_data_test)
        test_misclassified = num_misclassified(iris_target_test, pred)
        
        perf = perf.append({'n_neighbors': n_neighbors, 'weights': weights, 'train': train_misclassified, 'test': test_misclassified}, ignore_index=True)
        
```

```python
plt.figure(figsize=(12,8))

uniform = perf[perf.weights == 'uniform']
distance = perf[perf.weights == 'distance']

plt.plot(uniform.n_neighbors, uniform.train, label='weights=uniform, train')
plt.plot(uniform.n_neighbors, uniform.test, label='weights=uniform, test')

plt.plot(distance.n_neighbors, distance.train, label='weights=distance, train')
plt.plot(distance.n_neighbors, distance.test, label='weights=distance, test')

plt.xlabel('n_neighbors')
plt.ylabel('# misclassified')
plt.legend(loc=4)
plt.ylim([-25, 1600])
```

Some observations from this set of experiments.

1. The reason you can't see the training results using 'distance' for the neighbor weights is that the classifier is always perfect on the training set.
2. We can see that on the test set, 'uniform' and 'distance' weighting are equivalent, so there is no reason to prefer one method over the other for this data.
3. If we hadn't performed cross-validation, we might not have noticed this.  'distance' weighting performs perfectly (because of how it is defined) for the data it is trained on.  But this method of weighting is really no better than any other on unknown data.
4. My example didn't end up being a very good example of overfitting.  We do see that we need at least 9 or 10 or so n_neighbor cluster's to be created for the KNN classifier, but performance doesn't degrade because of overfitting if we use many more neighbor clusters, at least up to 50.  Though the performance of 'distance' weighting on the trained data, always being 0, is a kind of overfitting, that we wouldn't detect if we didn't do the cross-validation.
