Linear Regression
=================

Introduction
------------

In the abstract, regression is a very simple concept: you want to predict one set of numbers given another set of numbers.  For example, we have already seen predicting a persons weight based on their hight, but finding a best fit line through the data and using this as a model for any given unknown person's height to guess their weight.  Another example of a prediction that one might want to make, actuaries want to predict how long a person will live given their smoking habits and other factors (important for the insurance industry).

In general, we will call the numbers we're given the inputs (also known as the independent variables) and the numbers we want to predict are the outputs (the dependent variables).  The inputs are also often referred to as features or pictures in ML circles.

Loading data
=====================================================
We load the dataset 'diabetes' using the sklearn load function: 

```python
from sklearn import datasets
# Load the diabetes dataset
diabetes = datasets.load_diabetes()
```

The dataset consists of data (inputs or the prediction factors) and targets. Target tells us what is the desired output for specific example from data: 

```python
X = diabetes.data
y = diabetes.target
print X.shape
print y.shape
```

Splitting the data
==================
We want to split the data into train set and test set. We fit the linear model on the train set, and we show that it performs good on test set. This is a standard technique in ML when developing a model.  We first develop it and test it on data we've already collected by creating the model on part of the data, and testing the accuracy on the remaining data.  

Before splitting the data, we shuffle (mix) the examples, because for some datasets the examples are ordered.  If we didn't perform this shuffle, the training set and testing set could be totally different because they wouldn't have data representitive of all of the input, only a portion. 

Now we shuffle (the random_state parameter here basically acts like a seed, if we shuffle original data with this random_state we will get the same shuffle):


```python
from sklearn.utils import shuffle
X, y = shuffle(X, y, random_state=1)
print X.shape
print y.shape
```

Each example of data has 10 columns in total.

We want to work with 1-dim data because it is simple to visualize. Therefore select only one column, e.g column 2 and fit linear model on it:

```python
# Use only one column from data
print(X.shape)
X = X[:, 2:3]
print(X.shape)
```

Split the data into training/testing sets.  This is particularly easy to do in Python and high level scripting languages, as we simply take  slices of the data at some point to split into train/test sets.

```python
train_set_size = 250
X_train = X[:train_set_size]  # selects first 250 rows (examples) for train set
X_test = X[train_set_size:]   # selects from row 250 until the last one for test set
print(X_train.shape)
print(X_test.shape)
```

Split the targets into the same training/testing sets.  It is important to realize that the shuffle() we performed previously will keep the ordering in the X and y parameters consistent, so that the outputs in y are shuffled to the same corresponding rows as their inputs in X.

```python
y_train = y[:train_set_size]   # selects first 250 rows (targets) for train set
y_test = y[train_set_size:]    # selects from row 250 until the last one for test set
print(y_train.shape)
print(y_test.shape)
```

Now we can look at our train data. We can see that the examples appear to have a positive linear relation. 

Therefore, we can use linear model to make good classification of our examples.  We have plotted the train and test sets as separate colors, so that you can also confirm that the data has been shuffled.  If the data were still ordered, we would see the red and blue points separated in such a plot.


```python
plt.scatter(X_train, y_train, c='b')
plt.scatter(X_test, y_test, c='r')
plt.xlabel('Data')
plt.ylabel('Target');
```

Linear regression
=================
Create linear regression object, which we use later to apply linear regression on data

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
```

Fit the model using the training set

```python
regr.fit(X_train, y_train);
```

We found the coefficients and the bias (the intercept)

```python
print(regr.coef_)
print(regr.intercept_)
```

Now we calculate the mean square error on the training set and the test set

```python
# The mean square error
print("Training error: ", np.mean((regr.predict(X_train) - y_train) ** 2))
print("Test     error: ", np.mean((regr.predict(X_test) - y_test) ** 2))
```

Plotting data and linear model
==============================

Here we visualize the result of our linear regression.  We plot the training data as dots.  The line represents the predictions from the linear regression model that we found.

```python
# Visualises dots, where each dot represent a data exaple and corresponding teacher
plt.scatter(X_train, y_train,  color='black')
# Plots the linear model
plt.plot(X_train, regr.predict(X_train), color='blue', linewidth=3);
plt.xlabel('Data')
plt.ylabel('Target')
```

We do similar with test data, to show that out linear model is valid for the test set:

```python
# Visualises dots, where each dot represent a data exaple and corresponding teacher
plt.scatter(X_test, y_test,  color='black')
# Plots the linear model
plt.plot(X_test, regr.predict(X_test), color='blue', linewidth=3);
plt.xlabel('Data')
plt.ylabel('Target');
```
