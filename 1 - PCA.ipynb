Scatterplots
============

When our data has only 2 dimensions (Height and Weight, we will ignore Gender for the moment), we can use a scatter plot of the data to see if there is any relationship between the variables.

We'll first pull out the weights, like we did with the heights previously, so that we have the heights and weights each in separate arrays.

```python
weights = heights_weights['Weight']
plt.figure(figsize=(12,12))
plt.plot(heights, weights, '.k', mew = 0, alpha = .1)
plt.xlabel('height (in)')
plt.ylabel('weight (lbs)')
```

You should be already aware that there is a positive correlation between a person's height and their weight.  The shorter the person, the less they are likely to weigh, and the taller the more we expect them to weigh.  This is seen in a scatterplot when it appears there is a line that fits the data (with some noise) and the line has some slope, positive or negative.  In this figure, there is a positive slope, indicating a positive correlation between height and weight.  If the data were to be completely horizontal or vertical, this implies that there is no correlation between the 2 dimensions (and if you think about this a bit, you should be able to intuitively understand why this is true).

Lets quickly look at 2 functions to calculate the corelation and covariance of this data.

```python
heights.corr(weights)
```

```python
heights.cov(weights)
```

```python
Y = np.column_stack((heights, weights))
#print Y.shape
#Y[:10,:]
np.cov(Y, rowvar=False)
```

The correlation is close to +1, which is very high (correlation ranges from -1 to +1).  Covariance is another method that indicates the amount of variance that the variables have that influences each other (don't worry for now if you don't exactly know what covariance is, we will discuss it more later).  Covariance can be greater than 1, but 0 still indicates no covariance, and positive values indicate a positive covariance.  I have shown a second method of calculating covariance, using a method from the NumPy library.  Notice the off diagonal gives the same answer as before.  The diagnols are the covariance of each variable with itself.

Generating Data
===============

Lets explore scatterplots and data with 2 dimensions a bit more to make sure you understand what correlation means and looks like in a 2D scatterplot.  Instead of using actual data, lets generate data at random with known correlation/covariance properties.

We first create 2000 random two-dimensional data points. The data points are sampled from a multinomial normal distribution.  Notice we define a covariance matrix first, with covariance value of -2.2 between the 2 dimensions.  This means we want to generate data with a (smallish) negative covariance).  The data will have 2 columns, each with 200 values.  The mean of the first variable/dimension will be 1, and the second variable/dimension will have mean 2.  The variance of the randomly generated data is controlled by the size of the diagnol values in the covariance matrix.

```python
Cov = np.array([[2.9, -2.2], [-2.2, 6.5]])
X = np.random.multivariate_normal([1,2], Cov, size=2000)
```

Let's have a look at the raw data first...

```python
np.set_printoptions(4, suppress=True) # show only four decimals
print X[:10,:] # print the first 10 rows of X (from 0 to 9)
```

Do you see a relationship between the two columns? Tricky, I'd say. However, since the data is two-dimensional, we can plot the data, which allows us to see the relationship.

```python
plt.figure(figsize=(12,12))
plt.scatter(X[:,0], X[:,1])
plt.axis('equal') # equal scaling on both axis;
```

You should see though the data is roundish, it definitely has a negative slope (it is not horizontal or vertical).  We can also have a look at the actual covariance matrix (this should be close to the covariance matrix we asked the random function to generate the data from):

```python
print np.cov(X,rowvar=False)
```

Lets see what the data looks like if we generate some random data with a 0 covariance.  You should try this yourself with various (larger and small) negative and positive covariances, as well as 0, untill you understand the relationship of the corelation/covariance between 2 variables, and their shape in a scatterplot.

```python
Cov = np.array([[15.0, 0.0], [0.0, 2.0]])
X = np.random.multivariate_normal([1,2], Cov, size=2000)
plt.figure(figsize=(12,12))
plt.scatter(X[:,0], X[:,1])
plt.axis('equal') # equal scaling on both axis;
print np.cov(X,rowvar=False)
```

Running PCA
===========

We would now like to analyze the directions in which the data varies most. For that, we 

1. place the point cloud in the center (0,0) and
2. rotate it, such that the direction with most variance is parallel to the x-axis.

Both steps can be done using PCA, which is conveniently available in sklearn.

We start by loading the PCA class from the sklearn package and creating an instance of the class:

```python
from sklearn.decomposition import PCA
pca = PCA()
```

Now, `pca` is an object which has a function `pca.fit_transform(x)` which performs both steps from above to its argument `x`, and returns the centered and rotated version of `x`.

```python
Cov = np.array([[2.9, -2.2], [-2.2, 6.5]])
X = np.random.multivariate_normal([1,2], Cov, size=2000) # Set X random data back to the -2.2 covariance example
X_pca = pca.fit_transform(X)
```

```python
pca.components_
```

```python
pca.mean_
```

```python
plt.figure(figsize=(12,12))
plt.scatter(X_pca[:,0], X_pca[:,1])
plt.axis('equal');
```

The covariances between different axes should be zero now. We can double-check by having a look at the non-diagonal entries of the covariance matrix:

```python
print np.cov(X_pca, rowvar=False)
```

High-Dimensional Data
=====================

Our small example above was very easy, since we could get insight into the data by simply plotting it. This approach will not work once you have more than 3 dimensions, let's say we have the same data, but it is represented in four dimensions:

```python
np.random.seed(1)
# this basically takes the dot product between the original X, and a new randomly generated array.  You should explore what
# np.random.uniform and np.random.randint are doing here.
X_HD = np.dot(X,np.random.uniform(0.2,3,(2,4))*(np.random.randint(0,2,(2,4))*2-1))
# but the end result is a new data set with 4 columns rather than 2, but all 4 columns are just constant combinations of the
# original 2 columns (so there is really still only 2 dimensions of data in this new data set).
```

```python
np.random.uniform(0.2, 3, (2,4))
```

```python
np.random.randint(0, 2, (2,4))*2-1
```

```python
M = np.random.uniform(0.2, 3, (2,4))  * (np.random.randint(0, 2, (2,4))*2-1)
print M
```

```python
X.shape, M.shape
```

Lets look at the data again. First, the raw data:

```python
print X_HD[:10]
```

That one is more tricky. See anything? We can also try plot a few two-dimensional projections:

```python
plt.figure(figsize=(12,12))
for i in xrange(4):
    for j in xrange(4):
        plt.subplot(4, 4, i * 4 + j + 1)
        plt.scatter(X_HD[:,i], X_HD[:,j])
        plt.axis('equal')
        plt.gca().set_aspect('equal')
```

It is not easy to see that this is still a two-dimensional dataset! 

However, if we now do PCA on it, you'll see that the last two dimensions do not matter at all:

```python
X_HE = pca.fit_transform(X_HD)
print X_HE[:10,:]
```

Here it is easy to see, that the data is **still only two-dimensional**. Let's plot the two dimensions.

```python
plt.figure(figsize=(12,12))
plt.scatter(X_HE[:,0], X_HE[:,1])
plt.axis('equal')
plt.gca().set_aspect('equal')
```

Why does the result look differently than the two-dimensional data from which we generated it?

```python
plt.figure(figsize=(8,8))
for i in xrange(4):
    for j in xrange(4):
        plt.subplot(4, 4, i * 4 + j + 1)
        plt.scatter(X_HE[:,i], X_HE[:,j])
        plt.gca().set_xlim(-40,40)
        plt.gca().set_ylim(-40,40)
        plt.axis('equal')
        plt.gca().set_aspect('equal')
```

Dimension Reduction with PCA
============================

We can see that there are actually only two dimensions in the dataset. 

Let's throw away even more data -- the second dimension -- and reconstruct the original data in `D`.

```python
pca = PCA(1) # only keep one dimension!
X_E = pca.fit_transform(X_HD)
print X_E[:10,:]
```

Now lets plot the reconstructed data and compare to the original data D. We plot the original data in red, and the reconstruction with only one dimension in blue:

```python
X_reconstructed = pca.inverse_transform(X_E)
plt.figure(figsize=(12,12))
for i in xrange(4):
    for j in xrange(4):
        plt.subplot(4, 4, i * 4 + j + 1)
        plt.scatter(X_HD[:,i], X_HD[:,j],c='r')
        plt.scatter(X_reconstructed[:,i], X_reconstructed[:,j],c='b')
        plt.axis('equal')
```

## PCA on Images

In this final example, we use the $k$-Means algorithm on the classical MNIST dataset.

The MNIST dataset contains images of hand-written digits. 

Let's first fetch the dataset from the internet (which may take a while, note the asterisk [*]):

```python
from sklearn.datasets import fetch_mldata
from sklearn.cluster import KMeans
from sklearn.utils import shuffle
X_digits, _,_, Y_digits = fetch_mldata("MNIST Original").values() # fetch dataset from internet
X_digits, Y_digits = shuffle(X_digits,Y_digits) # shuffle dataset (which is ordered!)
X_digits = X_digits[-5000:]       # take only the last instances, to shorten runtime of PCA
```

Let's have a look at some of the instances in the dataset we just loaded:

```python
plt.rc("image", cmap="binary")
plt.figure(figsize=(8,4))
for i in xrange(10):
    plt.subplot(2,5,i+1)
    plt.imshow(X_digits[i].reshape(28,28))
    plt.xticks(())
    plt.yticks(())
plt.tight_layout()
```

**Warning**: This takes quite a few seconds, so be patient until the asterisk [*] disappears!

```python
from sklearn.decomposition import PCA
pca = PCA()
X2_digits = pca.fit_transform(X_digits)
```

It does not make much sense to look at the transformed images, they will look like noise to us. 

Instead, let's have a look at the most important directions on which the dataset was projected:

```python
plt.figure(figsize=(8,6))
W = pca.components_

for i in xrange(10): # loop over all means
    plt.subplot(2,5,i+1)
    plt.imshow(W[i].reshape(28,28))
    plt.xticks(())
    plt.yticks(())
plt.tight_layout()
```

The later directions (here, from the 100-th on) mainly show noise, small variations between different, but very similar training instances:

```python
plt.figure(figsize=(8,4))
W = pca.components_

for i in xrange(10): # loop over all means
    plt.subplot(2,5,i+1)
    plt.imshow(W[100+i].reshape(28,28))
    plt.xticks(())
    plt.yticks(())
plt.tight_layout()
```

The number of "interesting" dimensions can be seen from the importance of the found directions. 

We can simply plot them:

```python
plt.plot(pca.explained_variance_);
```

We can see that the intrinsic dimensionality is not higher than maybe 100, even though the dataset has 784 dimensions!

Let's reconstruct the data again using only a handfull of the 784 dimensions:

```python
from sklearn.decomposition import PCA
pca = PCA(20)
X2_few_digits = pca.fit_transform(X_digits)
```

```python
plt.figure(figsize=(16,4))
X_recons_digits = pca.inverse_transform(X2_few_digits)
for i in xrange(10):
    plt.subplot(2,10,i+1)
    plt.imshow(X_recons_digits[i].reshape(28,28))
    plt.xticks(())
    plt.yticks(())
for i in xrange(10):
    plt.subplot(2,10,11+i)
    plt.imshow(X_digits[i].reshape(28,28))
    plt.xticks(())
    plt.yticks(())
plt.tight_layout()
```
