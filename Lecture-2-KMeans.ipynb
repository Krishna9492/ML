The $k$-Means Algorithm
====================

Again, we start by generating some artificial data:

```python
plt.jet() # set the color map. When your colors are lost, re-run this.
import sklearn.datasets as datasets
X, Y = datasets.make_blobs(centers=4, cluster_std=0.5, random_state=0)
```

As always, we first *plot* the data to get a feeling of what we're dealing with:

```python
plt.scatter(X[:,0], X[:,1]);
```

The data looks like it may contain four different "types" of data point. 

In fact, this is how it was created above.

We can plot this information as well, using color:

```python
plt.scatter(X[:,0], X[:,1], c=Y);
```

Normally, you do not know the information in `Y`, however.  Lets look at what is in Y:

```python
print X[:5,]
print Y[:5,]
```

Y is just a label (an integer 0,1,2 or 3) that indicates the group each point in X belongs to.  

We may not have the labels in Y, but we can try to recover or determine them from the data alone.

This is what the kMeans algorithm does.  

```python
from sklearn.cluster import KMeans
kmeans = KMeans(4, random_state=8)
Y_hat = kmeans.fit(X).labels_
```

Now the label assignments should be quite similar to `Y`, up to a different ordering of the colors:

```python
plt.scatter(X[:,0], X[:,1], c=Y_hat);
```

Often, you're not so much interested in the assignments to the means. 

You'll want to have a closer look at the means $\mu$.

The means in $\mu$ can be seen as *representatives* of their respective cluster.

```python
plt.scatter(X[:,0], X[:,1], c=Y_hat, alpha=0.4)
mu = kmeans.cluster_centers_
plt.scatter(mu[:,0], mu[:,1], s=100, c=np.unique(Y_hat))
print mu
```

## $k$-Means on Images

In this final example, we use the $k$-Means algorithm on the classical MNIST dataset.

The MNIST dataset contains images of hand-written digits. 

Let's first fetch the dataset from the internet (which may take a while, note the asterisk [*]):

```python
from sklearn.datasets import fetch_mldata
from sklearn.cluster import KMeans
from sklearn.utils import shuffle
X_digits, _,_, Y_digits = fetch_mldata("MNIST Original").values() # fetch dataset from internet
X_digits, Y_digits = shuffle(X_digits,Y_digits) # shuffle dataset (which is ordered!)
X_digits = X_digits[-5000:]       # take only the last instances, to shorten runtime of KMeans
```

Let's have a look at some of the instances in the dataset we just loaded:

```python
plt.rc("image", cmap="binary") # use black/white palette for plotting
for i in xrange(10):
    plt.subplot(2,5,i+1)
    plt.imshow(X_digits[i].reshape(28,28))
    plt.xticks(())
    plt.yticks(())
plt.tight_layout()
```

Lets use $k$-Means to find clusters in the data.  We will ask it to create 20 clusters from the data.

**Warning**: This takes quite a few seconds, so be patient until the asterisk [*] disappears!

```python
kmeans = KMeans(20)
mu_digits = kmeans.fit(X_digits).cluster_centers_
```

Let's have a closer look at the means. Even though there are 10 digits, some of them are over/under-represented. Do you understand why?

```python
plt.figure(figsize=(16,6))
for i in xrange(2*(mu_digits.shape[0]/2)): # loop over all means
    plt.subplot(2,mu_digits.shape[0]/2,i+1)
    plt.imshow(mu_digits[i].reshape(28,28))
    plt.xticks(())
    plt.yticks(())
plt.tight_layout()
```

Example implementation of K-means clustering in Python
======================================================

Lets start by creating a simple function to compute distance between 2 n-dimensional points
Points are assumed to be lists or NumPy 1-dimensional arrays (vector)

```python
def euclidean_distance(p1, p2):
    """Calculate the euclidean distance from p1 to p2.  We demonstrate an iterative
    method here (e.g. non-vectorized), so implementation will not be particularly efficient.

    pt1, pt2  n-dimensional points, must be list-like (conform to List semantics)

    returns distance"""
    d = 0.0
    # first calculate the sum of the squared differences
    for x1,x2 in zip(p1,p2):
      d += (x1 - x2) ** 2.0
    # then distance is the square root of this sum squared value
    return d**0.5
```

Always a good idea to test our code a bit.  Lets make sure euclidean_distance works on some examples.

```python
p1 = np.array([0, 0])
p2 = np.array([0, 1])
print euclidean_distance(p1,p2)
p1 = np.array([0, 0])
p2 = np.array([1, 1])
print euclidean_distance(p1,p2)
p1 = np.array([0, 0, 0, 0, 0])
p2 = np.array([1, 1, 1, 1, 1])
print euclidean_distance(p1,p2)
```

For performing K-means clustering, we will be given a d x N dimensional NumPy array.  Each point has d dimensions (the columns of the data array).  The data set consists of N points of data.  Lets generate a
set of clusterd points.

In the following demonstration, we generate a set of clustered data that has 2 dimensions (n_features).  We generate 1000 points of data.  The data will be clustered into 4 clusters, where each cluster has a std of 0.5.

```python
import sklearn.datasets as datasets
X, Y = datasets.make_blobs(n_samples=1000, n_features=2, centers=4, cluster_std=0.5, random_state=0)
print X.shape # X has 1000 rows or items of data, with 2 dimensions or features for each data point
print X[:10,:] # display first 10 2-d points in X
```

So we have our set of items we want to cluster in X.  The algorithm for K-means clustering requires that we generate a set of centroids at random locations.  For example, if we want the algorithm to discover 4
clusters for use, we need to start by generating 4 points at random.  The 4 points should have the same number of dimensions as our dataset X (2 dimensions in this case).  Further, it is best if the 4 points are within the range of each dimension.

For example, for X, all of the points in the first dimension can be found like this:

```python
print X[:10,0] # print first 10 points of the first (0th indexed) dimension
```

In order go generate a point that is within the correct range, we need to know the min and max value that occurs in each dimension of the dataset X.  For the 0th dimension, we can use the min and max functions to
determine this.

```python
print min(X[:,0]), max(X[:,0]) # the minumum and maximum values of the 0th dimension, e.g. the range
print min(X[:,1]), max(X[:,1]) # same for the 1th indexed dimension
```

Lets save these ranges in a variable, as we will need them.  Here is an example of using what are known as list productions to turn this into a one liner which will work for a dataset X that has any number of features or dimensions

```python
ranges = [ (min(X[:,di]), max(X[:,di])) for di in range(X.shape[1])]
print ranges
```

So here we have created a new list of tuples, called ranges.  The 0th item in ranges contains the min and
max values for the 0th dimension of X, and so on.

```python
print "Minimum of 0th dimension: ", ranges[0][0]
print "Maximum of 0th dimension: ", ranges[0][1]
```

Given ranges, we can perform the initialization step of the K-means algorithm.  In this example we have created data that has 4 clusters.  We therefore want to run K-means and ask it to form or discover 4 clusters.  We need to start by creating 4 points at random (called centroids).  These points should have values that are all within the ranges of the dataset X on all dimensions.  So for example, to create
1 point with 2 dimensions within the ranges of the dataset X we could do this:

```python
c1 = [random.random() * ((ranges[0][1] - ranges[0][0]) + ranges[0][0]),
      random.random() * ((ranges[1][1] - ranges[1][0]) + ranges[1][0])]
```

```python
print c1
```

The previous code first generates a random number between 0 and 1.  It then multiplies it by the difference of the max from min for the dimension (this is known as the span of the dimension.  And we then add on the min.  The result is a random number between the minum and maximum for each dimension (you should convience yourself of this fact).

We can use a list production like we did to create the ranges in order to generate all 4 starting centroid points at 1 time:

```python
k = 4 # number of centroids to generate at random / number of clusters we will discover
d = X.shape[1] # number of features or dimensions of X data set
# in order to be general, we need to handle any requested value of k, but also data sets with
# different numbers of dimensions or features
centroids = [[random.random() * ((ranges[i][1] - ranges[i][0]) + ranges[i][0]) 
                for i in range(d)] for j in range(k)]
print centroids
```

K-means is a stochastic algorithm, because we choose the starting centroid values at random as we have
just done.  K-menas is also an iterative algorithm, we keep iterating over some operations until we detect that our cluster formations meet some criteria of goodness.

The first thing we need to do, is be able to calculate the distance from every point in X to every one of the centroids.  We then need to see which of the 4 centroids is closest to each point in X, and remember which centroid each point is associated with.

We will start by writing a function that takes the dataset X and the centroids, and returning a list of the distances between each X and each of the points in the centroids.

```python
def distance_to_centroids(dataset, centroids):
    """Given a dataset of points, and a set of centroids, return the distance of each point in
    the dataset to each of the centroids."""
    # we will return the result in a list called distances
    distances = [] 
    
    # look at each point p, 1 at a time, in the dataset
    for p in dataset: 
        distances_for_p = []
        # calculate distance from p to each centroid
        for c in centroids:
            d = euclidean_distance(p, c)
            distances_for_p.append(d)
        distances.append(distances_for_p)
        
    return distances
            
```

```python
distances = distance_to_centroids(X, centroids)
print distances[:10]
```

distances now contains a nested list.  Each item in the list contains a list of 4 distances, the distance from the corresponding point to each of the centroids.

We now need to determine which of the 4 centroids each point is closest to.  For example, we can find which of the 4 centroids (0-3) the first point is closest to like this:

```python
distances[0].index(min(distances[0]))
```

Here the index method of a list returns the index of the value you pass to it.  So in english we basically ask what is the index of the minimum value for the 0th point, where this index will correspond to the centroid number.  We can determine the closest centroid for all points using a 1 liner list comprehension:

```python
bestmatches = [ distances[i].index(min(distances[i])) for i in range(len(distances)) ]
print bestmatches[:10]
```

There is basically on more piece then to perform the K-means clustering algorithm.  Given the best matches of each point to the current set of centroids, we need to calculate a new set of (4) centroids.  The new centroid will be the at the center of all the points associted with that centroid.  For example, this means we want to find all of the bestmatches that are close to the centroid 0, then calculate a new centroid 0 as the centroid of these points.  A centroid is very easy to calculate, even in n-dimensional space, it is the average of all the points (sum up the values for each dimension, then calculate the average).

So, lets calculate the new centroid for points closest currently to centroid 0:

```python
num_matches = 0
new_centroid = [0.0, 0.0] # we will first use to sum up all the dimensions, then find the average
# start by searching for all of the bestmatches to centroid 0
for i in range(len(bestmatches)):
    if bestmatches[i] == 0:
        num_matches += 1
        new_centroid[0] += X[i, 0] # dimension 0
        new_centroid[1] += X[i, 1] # dimension 1
        
# finish by calculating the average in each dimension
new_centroid[0] = new_centroid[0] / float(num_matches)
new_centroid[1] = new_centroid[1] / float(num_matches)

print new_centroid
```

It will be hard to tell if this is working or not at this point, though you should see that the new_centroid has moved, it will be different from the original 0th centroid.  Of course the above code will need to be made general.  We need to calculate all k (4 in our example) new centroids, which we could do all in the same loop.  We also need to make sure we are handling n-dimensions (not just our example 2 dimensions).

Lets write a general function that will take the bestmatches, and the dataset (X), and return the calculated set of new_centroids, that will work for any k and any number of dimensions.

```python
def move_centroids(bestmatches, dataset, k):
    """Given a dataset and the current list of which centroid each dataset point is matched
    with (as well as k, the number of centroid clusters being computed),
    return a new set of centroids calculated as the average of the current group of 
    best matching points"""
    d = dataset.shape[1] # number of features/dimensions
    # will hold the running sums, and eventually the averaged new centroids to return
    centroids = [ [ 0.0 for i in range(2) ] for j in range(4)]
    num_matches = [0.0] * k # number of items we find associated with a centroid
    
    # look at each point i and centroid c it is associated with
    for i,c in enumerate(bestmatches):
        num_matches[c] += 1.0
        for dn in range(d):
            centroids[c][dn] += dataset[i, dn]
    
    # now calculate averages
    for c in range(len(centroids)):
        for dn in range(d):
            if num_matches[c] > 0.0: # watch out for divide/by/zero, no points associated w/ centroid
                centroids[c][dn] /= num_matches[c]
            else:
                centroids[c][dn] = 0.0
            
    return centroids
```

```python
new_centroids = move_centroids(bestmatches, X, 4)
print new_centroids
```

We are finally ready to put all of this together.  The only thing that is missing is that, after creating a set of new_centroids, we need to see if the centroids have stopped moving.  We can do this in 2 ways, by comparing the old and new centroids, or we could see if the old bestmatches and new bestmatches have changed.

If either of these are different, then it means we are not done yet.  If we are not done, we need to loop back and calculate bestmatches and move/create a new set of centroids.

Here is a complete implementation of K-means, using all of the previous pieces, to accomplish this task.

```python
def kmeans(dataset, k=4):
    """We expect the dataset to be a NumPy array, with data arranged in rows, and the columns
    with measures of the features or different dimensions of the dataset.  We will cluster the
    items into k clusters (default is 4).

    We return a list of the categories each data item in datasets is associated with.  This is
    an integer number that ranges from 0 to k-1"""

    d = dataset.shape[1] # number of features/dimensions
    
    # determine the min/max e.g ranges for each dimension
    ranges = [ (min(dataset[:,di]), max(dataset[:,di])) for di in range(X.shape[1]) ]
    
    # create k randomly placed initial centroids, the stochastic part of the algorithm
    centroids = [[random.random() * ((ranges[i][1] - ranges[i][0]) + ranges[i][0]) 
                    for i in range(d)] for j in range(k)]
    
    lastmatches = None # used to determine when we are done iterating
    for t in range(100): # we will iterate till done, or a maximum hardcoded of 100 iterations
        print 'Iteration %d' % t
        
        # Find which centroid is the closest for each row
        distances = distance_to_centroids(dataset, centroids)
        bestmatches = [ distances[i].index(min(distances[i])) for i in range(len(distances)) ]
        
        # If the results have not changed from last iteration, then we are done
        if bestmatches == lastmatches:
            break
        lastmatches = bestmatches
        
        # move the centroids to the average of their members
        centroids = move_centroids(bestmatches, dataset, k)
        
    return bestmatches
```
