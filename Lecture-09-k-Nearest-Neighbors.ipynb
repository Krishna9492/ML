k Nearest Neighbors
-------------------

**Building an Artificial Model**

In this lecture we will look at basic classification technique, known as k Nearest Neighbors (or kNN).  Do not confuse kNN with
the KMeans clustering technique we previously looked at.  KMeans clustering is an unsupervised method, used to discover natural
clusters, or groupings, among a data set.  kNN is an example of a classification technique, and is is a supervised learning
technique.  In kNN, we have a set of known data, that has some attribute we would like to predict for similar but unknown data.
However, unlike some other classification methods we will look at, kNN can be used to do classification on continuous data, e.g.
on data that is not some set of predefined categories, but instead can take on any value within some range.  For example, if
we want to predict the price of something, say a house we wish to buy, based on some features, the number of rooms, square
footage, location, etc., then we need a classification or prediction technique that can be used to predict a price value of the
house, which can take on any dollar amount that the market deems appropriate for the house with the given features.

Lets make up another example data set using price to make this clearer.  Lets say that we want to build a model to predict the
price of a bottle of wine.  Lets make an artificial model of wine price that we will then use kNN to try and model and predict
our artifical ground truth we create.  For example, we could say that the price of a bottle of wine is determined by a
combination of factors based on the rating of the wine and its age.  Our artificial model will assume that each wine has a 
peak age, which is older for good wines (with higher ratings), and almost immediate for bad wines.  A high-rated wine will
start at a high prices and increase in value until its peak age.  A low-rated wine will start cheap and get cheaper after
almost immediately passing its peak age.

Lets create a function that defines this artificial model, and basically defines the "ground truth" of the system we will try
to build our kNN classifier for.

```python
def wineprice(rating, age):
    """Determine the price of a wine, given the wines rating (a measure of its quality), and its current age.
    The rating, in our simple model of wine quality/pricing, directly corresponds to the peak age of the wine,
    higher rated wines have a longer and older peak age, lower rated wines have a short peak age.  
    
    rating - A scalar float, a value between 51 and 100
    age - A scalar float or int, the age of the wine in years
    
    returns - the wine price in US dollars, always > $0.00
    """
    # 50 is lowest quality/rating, and peaks at age 0, 100 is highest quality, and peaks in 50 years
    peak_age = float(rating) - 50.0 
 
    # factor to increase/decrease base price up to at peak age
    mult = peak_age / 5.0
    if mult < 1.0:
        mult = 1.0

    # calculate price based on rating, the base price is also a function of the rating
    price = float(rating) / 4.0
    if age > peak_age:
        # past its peak, goes bad in 5 years
        price = price * (mult - (age - peak_age))
    else:
        # increase to some multiple of original value as it approaches peak age
        price = price * (mult * ((age + 1) / peak_age))
        
    if price < 0:
        price = 0.0
        
    return price
```

```python
print wineprice(51, 1)
```

Just to get a feel for our artificial model, lets plot the price of wines of different qualities, as they age from 0 to 60
years.

```python
plt.figure(figsize=(12,8))

wine_ratings = [51, 60, 70, 80, 90, 100]
ages = range(1, 61)
for rating in wine_ratings:
    prices = []
    for age in ages:
        prices.append(wineprice(rating, age))
    
    l = "%d" % rating
    plt.plot(ages, prices, label=l) 
    
plt.legend()
plt.xlabel('wine age (years)')
plt.ylabel('wine price ($ USD)')
```

So basically, the higher the quality the wine, the higher its price as it approaches its peak age.  All wines decrease in
value to $0.00 after 5 years beyond their peak age, and slowly increase in value up to their peak price as they
approach their peak age.

We can use this artifical model of reality to create a data set of wine prices.  For example, lets create a numpy array representing
200 bottles of wine.  All of the ages should be between 1 and 60 years of age, and all of the qualities should be between 51 and 100
quality rating points.  We will calculate the price for the given randomly generated rating/age.  These 3 attributes will make up
3 columns of our data set, of which the first 2 columns (rating, age) are the independent measures we are trying to build a 
prediciton model of, and the wine prices is our third column represent our dependent measure we are trying to predict.

```python
NUM_BOTTLES = 200
ratings = np.random.uniform(51.0, 100.0, size=(NUM_BOTTLES))
ages = np.random.uniform(1.0, 60.0, size=(NUM_BOTTLES))
prices = np.zeros((NUM_BOTTLES))
for i, (rating, age) in enumerate(zip(ratings, ages)):
    prices[i] = wineprice(rating, age)
bottles = np.zeros((NUM_BOTTLES, 3))
bottles[:,0] = ratings
bottles[:,1] = ages
bottles[:,2] = prices
print bottles[:5,:]
```

Here bottles is a randomly generated data set of 200 bottles of wine.  Each bottle is a row in the data set.  There are 3 columns
which are the 2 features plus 1 lable of our data.  The columns 0 and 1 are the wine rating and age respectively, while column
2 is the price of the wine.

In order to make this a slightly more realistic data set, we will add some random variation/noise to our artificial data set.  Lets
randomly decrease/increase the price from 80% to 120% of the original price (to reflect mark up, imperfect market information, etc.)


```python
noise = np.random.uniform(0.8, 1.2, size=(NUM_BOTTLES))
bottles[:,2] = noise * bottles[:,2]
print noise[:5]
print bottles[:5, :]
```

**k-Nearest Neighbors (kNN)**

kNN is conceptually fairly simple in how it predicts or classifies the data it is asked to model.  It is essentially the same
procedure as you might do when trying to determine what price you should pay for some item you are interested in purchasing.
You might search for similar items (similar in features and quality) from different sources, and then look at the prices asked
for the similar items you find.  Then, you would determine the price you ought to pay based on the price of other similar
items.  For the artificial wine price model we created, we might want to find wines of similar quality and age to a wine we
are interested in buying, and predict that its price should be similar to the other wines of similar quality and age that
we discovered.

The kNN classifier does basically this.  It finds a set of "similar" items to the one we are trying to predict, and then it 
makes a prediction based on the labeled attribute of the similar items it is basing its prediction on.  The trick is defining
a definition of "similar" that can be operationalized so that a machine can calculate it.  One easy way to define
similarity is to visualize the attributes of our data (the wine rating and age for our artificial model) as points in
a multi-dimensional space.  Then similarity can be measured simply as the distance between points, 2 points that are closer
together are more similar in their features/attributes.  For example, since our wine price system has only 2 attributes, we
can visualize all of our data as a 2 dimensional system:

```python
plt.figure(figsize=(8,8))
plt.plot(bottles[:,0], bottles[:,1], 'ko')
plt.xlabel('wine rating')
plt.ylabel('wine age')
```

Here we show all 200 bottles of our wine in our data set as points in a 2-dimensional space.  Points that are close together in
this space represent bottles of wine that have both similar ratings and ages.  

As another example, we can use simple Euclidean distance to formalize and calculate the "similarity" or distance between
bottles of wine in our feature/attribute 2-dimensional space.  Say for example we were interested in a bottle of wine
with a rating of 80, and an age of 25:

```python
plt.figure(figsize=(8,8))
plt.plot(bottles[:,0], bottles[:,1], 'ko')
plt.xlabel('wine rating')
plt.ylabel('wine age')

plt.plot(80.0, 25.0, 'rd')
```

We could calculate the distance between this wine with an rating of 80.0 and an age of 30.0 and all of our other wines, sort
the results by distance (similarity) and see the k closest neighbors to this wine (for say k=3, the three closest neighbors),
like this:

```python
# calculate distance between the wine we are trying to predict, and all others, using Euclidean distance
distance = ( (bottles[:,0]-80.0)**2.0 + (bottles[:,1]-25.0)**2.0 )**0.5
idx = np.argsort(distance) # returns indexes that would sort data in ascending order
# for k=5 nearest neighbors, use index to find 5 closest neighbors to bottle we are trying to predict
k = 5
print bottles[idx[:k], :]
print idx[:k]
print distance[idx[:k]]
```

We can also visualize/plot this k=5 nearest neighbors that were found

```python
plt.figure(figsize=(8,8))
plt.plot(bottles[:,0], bottles[:,1], 'ko')
plt.xlabel('wine rating')
plt.ylabel('wine age')

plt.plot(80.0, 25.0, 'rd')
plt.plot(bottles[idx[:k],0], bottles[idx[:k],1], 'ro')
```

One common way to make a prediction using kNN on continuous attribute data is to simply average the predicted property of the k
neighbors we determined were most similar.  So we could average the price of these 5 bottles of wine, and then see how close
that prediction is to the artifical model price, like this:

```python
# average the k=5 neareast neighbors price
print "k=5 nearest neighbors average price:", bottles[idx[:k], 2].mean()

# see what the price of a bottle of wine is from our artificial model
print "artificial model actual price:", wineprice(80.0, 25.0)
```

Did the kNN model do well in predicting the price?  Since we randomly generate our data, you can see big differences in the
performance.  Also, I picked a bottle of wine that was more than 5 years before its peak age.  But prices can fluctuate much
more rapidly after the peak age, and of course they quickly drop from very high prices to $0.0 for highly rated
wines (in our artificial model), which can make prediciton very hard, and very dependent on the random bottles of
wine you have sampled in your data set.  

Also, why did we pick k=5 neighbors?  The number of neighbors you should choose can greatly affect the accuracy of your
kNN prediction results.  Pick too few neighbors, and you are subject to the randomness or noise in the data of a single or
very few points.  But use too many neighbors and you will be using items that are not very similar to the one you are trying
to predict, and thus will also get bad results.  Choosing the right size for k is as much an art as a science, there is not
method to determine the optimum value for k when using kNN.  Also, you should be aware that there are other definitions of
"similarity" that can be defined and used when performing kNN.  Some measure of similarity will work better than others for
different types of data.  Another refinement that kNN models can support is to weight the neighbors based on their similarity
or closeness to the point you are trying to predict.  When using weighting, you could for example use more neighbors, but
less similar neighbors would be weighted less in producing the prediction for an input.

SKLearn kNN
-----------

Lets now show how to use the `sklearn` library to perform a kNN modeling and prediction for us.  As you would expect, `sklearn` 
supports all of the different concepts of similarity, different weighting methods, and other refinements when defining
and using kNN to model and predict a set of data.  As you have seen before, the pattern to use `sklearn`
for kNN fitting and prediction is the same as before.  We can use k=5 neighbors to do the same prediction on our
bottles of wine data set like this:


```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)    # we specify that this knn should always use 5 neighbors
features = bottles[:,0:2]
prices = bottles[:,2]
knn.fit(features, prices)

print knn.predict([80.0, 25.0])
print prices[idx[:5]]
```

But whoops, we didn't get the same result.  That is because the standard KNeighborsClassifier is supposed to be used for a 
standard classification task, where we have discrete labels, and we determine the predicted label based on some weighting of the
labels of the neighbors.  However, if we want to use kNN to predict a continuous value, this is actually a regression problem,
as we have seen for linear regression.  SKLearn does support regression prediction using k-Nearest Neighbors, but we need to use
the KNeighborsRegressor class instead:

```python
from sklearn.neighbors import KNeighborsRegressor
knn = KNeighborsRegressor(n_neighbors=5)    # we specify that this knn should always use 5 neighbors
features = bottles[:,0:2]
prices = bottles[:,2]
knn.fit(features, prices)

print knn.predict([80.0, 25.0])
print prices[idx[:5]]
```

Using the KNeighborsRegressor we get exactly the same result as when we did it by hand.  This is because the default is to
use uniform weighting to average the neighbors found to the item we are trying to predict.  And it uses Euclidean distance
as the measure of similarity by default.

Circles

We create a dataset consisting of a small circle surrounded by a larger circle.

First, we will try to fit is using logistic regression, than we will use $k$ Nearest Neighbors.


```python
from sklearn.datasets import make_circles
X, y = make_circles(noise=.1, factor=.5)
print "X.shape:", X.shape
print "unique labels: ", np.unique(y)
```

Now let's plot the data again.

```python
plt.prism()  # this sets a nice color map
plt.scatter(X[:, 0], X[:, 1], c=y)
```

Take the first 50 examples for training and the rest for testing.

```python
X_train = X[:50]
y_train = y[:50]
X_test = X[50:]
y_test = y[50:]
```

Import logistic regression and fit the model.

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
```

Evaluate the logistic regression as we did before by ploting the decision surface and predictions on the test data.
We plot the training data as circles, colored with their true labels.
The test data is colored with their prediction and plotted as triangles.

```python
plt.prism()
from utility import plot_decision_boundary
y_pred_test = logreg.predict(X_test)
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_pred_test, marker='^')
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train)
plot_decision_boundary(logreg, X)
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
print "Accuracy of logistic regression on test set:", logreg.score(X_test, y_test)
```

That doesn't look as good as before. Notice that all test points on the right of the line are red and all on the left are green. (as expected)

Now let us look at how K Nearest Neighbors works here. Let us import the classifier and create an object.

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)    # we specify that this knn should always use 5 neighbors
```

```python
knn.fit(X_train, y_train)
y_pred_test = knn.predict(X_test)
```

```python
plt.prism() # gives us a nice color map
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_pred_test, marker='^')
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train)
print "Accuracy of KNN test set:", knn.score(X_test, y_test)
```

This looks much better, which is also reflected by the test set score. You can try to change $k$ to make the prediction better.



MNIST
-----
We'll now have a look at MNIST again.

```python
from sklearn.datasets import fetch_mldata
from sklearn.utils import shuffle
mnist = fetch_mldata("MNIST original")
X_digits, y_digits = mnist.data, mnist.target
X_digits, y_digits = shuffle(X_digits, y_digits)
```

This time we use all classes, but only a small training set (because KNN usually takes a while).
To do model selection, we also create a valdation set to adjust $k$.

```python
X_digits_train = X_digits[:1000]
y_digits_train = y_digits[:1000]
X_digits_valid = X_digits[1000:2000]
y_digits_valid = y_digits[1000:2000]
X_digits_test = X_digits[2000:3000]
y_digits_test = y_digits[2000:3000]
```

Now let us fit the model. That actually just remembers the dataset. Then we will evaluate on the validation set.

This time we choose $k=20$.
You can find a good value later.


```python
knn_digits = KNeighborsClassifier(n_neighbors=20)
knn_digits.fit(X_digits_train, y_digits_train)
print "KNN validation accuracy on MNIST digits: ", knn_digits.score(X_digits_valid, y_digits_valid)
```

After you found a good value of $k$, you can evaluate again on the test set (only do this once to have a meaningful result!)

```python
print "KNN test accuracy on MNIST digits: ", knn_digits.score(X_digits_test, y_digits_test)
```

To get a better understanding of the classifier, let us take a closer look at some mistake that are done with $k=3$.

```python
knn_digits = KNeighborsClassifier(n_neighbors=3)
knn_digits.fit(X_digits_train, y_digits_train)
y_digits_valid_pred = knn_digits.predict(X_digits_valid)
```

Get the neighbors of the validation data from the training data.

```python
neighbors = knn_digits._tree.query(X_digits_valid, k=3)[1]
```

Not let's look at them. Let's start with an image where it worked. First plot the validation image itself, then three neighbors.

```python
plt.rc("image", cmap="binary")  # this sets a black on white colormap
# plot X_digits_valid[0]
plt.subplot(1, 4, 1)
plt.imshow(X_digits_valid[0].reshape(28, 28))
plt.title("Query")
# plot three nearest neighbors from the training set
for i in [0, 1, 2]:
    plt.subplot(1, 4, 2 + i)
    plt.title("%dth neighbor" % i)
    plt.imshow(X_digits_train[neighbors[0, i]].reshape(28, 28)) 
```

Find out where we went wrong on the validation set, so we can have a look.

```python
wrong = np.where(y_digits_valid_pred != y_digits_valid)[0]  # the != part gives a mask, the "where" gives us the indices
print "Wrong prediction on the following images: ", wrong

```

Now take one of these and visualize the 3 closest neighbors.
That will hopefully give us some insights into why there was an error.

```python
index = wrong[0]

plt.rc("image", cmap="binary")  # this sets a black on white colormap
# plot X_digits_valid[index]
plt.subplot(1, 4, 1)
plt.imshow(X_digits_valid[index].reshape(28, 28))
plt.title("Query")
# plot three nearest neighbors from the training set
for i in [0, 1, 2]:
    plt.subplot(1, 4, 2 + i)
    plt.title("%dth neighbor" % i)
    plt.imshow(X_digits_train[neighbors[index, i]].reshape(28, 28))
```
