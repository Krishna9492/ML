# Classification

Classification is a supervised learning technique.  In a typical classification task, we have a set of data where some features represent the inputs of the problem, and usually 1 feature reqpresents a discrete class that the individual observations has been classified in.  The problem in a classification task then becomes one of building a model from the given data that successfully predicts the assigned class.  The techniques used in classification are considered supervised, because they start with a set of pre-labeled data from which to form the model.  The classification model can then be measured on how successful it is with unlabled and previously unseen data.

A Simple Classifiction using Linear Regression
----------------------------------------------

To begin our discussion of classification, lets return to our trusty heights and weights data set.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
!head data/01_heights_weights_genders.csv
```

Notice that column 0 of our data set, `Gender`, is really a discrete binary class, of the type we were just talking about.  It takes on one of two values: `Male` or `Female`

```python
hwg = pd.read_csv('data/01_heights_weights_genders.csv')
height = hwg.Height.values
weight = hwg.Weight.values
male_indxs = hwg.Gender=='Male'
female_indxs = hwg.Gender=='Female'
```

```python
hwg
```

We haven't made use of it before, but the height/weight data set actually contains a third factor, the gender.  Lets look at our data again, but add in extra information about the gender of each individual.  (We are actually working in 3 dimensions now: height, weight and gender.

```python
plt.figure(figsize=(12,12))
plt.plot(height[male_indxs], weight[male_indxs], '.', label='Male', alpha=0.15)
plt.plot(height[female_indxs], weight[female_indxs], '.', label='Female', alpha=0.15)
plt.legend()
plt.xlabel('Height (in)')
plt.ylabel('Weight (in)')
plt.grid('on')
```

As you can see from this figure, there is some overlap between genders (I've used a bit of alpha so that the points are partially transparent, to help see some of the overlapping points).  So if we had a new data point, say someone with a hight of 65in and a weight of 125 lbs, we would be very confident in a prediction that the person is probably female.  It gets tougher for points near the boundary, but it appears that we could draw a line through this figure, separating the values into two classes.  New data falling on one side of the line would be predicted to be for a Male, and on the other a Female, and we would expect to be correct very often just looking at this figure.

Lets formalize this a bit.  We can use the simple linear regression from the previous lecture in order to find a line that best 'separates' this data.  We do this by using the height and weight as the 2 prediction input factors, and we create a best-fit linear hyperplane that best separates the values into the factor we are trying to predict, gender in this case.

Linear regression only works with numeric data.  In this case, the gender factor from our data set is actually a categorical variable, it only takes on 2 distinct values (`Male` or `Female`).  It is often the case for classifications that we will work with categorical variables like this, and try and build a model that correctly classifies (or categorizes if you like) unseen values into the most likely category.  However, many machine learning algorithms are mathematical in nature, and conceive of and manipulate the data points as an N-dimensional space.  This is easy to overcome, in this case, for linear regression.  We can simply turn the gender category into a numeric value appropriate for linear regression to work with.  We will  use 0.0 to indicate Male, and 1.0 to indicate Female (as we did with the smokers/nonsmokers example in previous lecture).

```python
gender = (hwg.Gender == 'Female').values * 1.0
gender = gender.reshape(gender.shape[0], 1) # make correct shape to make linear_model happy
# while were at it, we'll pull out height and weight into a 2 by N array that linear_model expects
factors = hwg[ ['Height', 'Weight'] ].values
print gender.shape
print factors.shape
```

Now we can perform a linear regression as in the previous section, with 2 factors to predict the gender.

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
regr.fit(factors, gender)
```

As before, the result is a hyperplane that best divides the points into the 2 categories (Male and Female).  The hyperplane is returned as an intercept and 2 slopes that define the plane.

```python
print "Slope = ", regr.coef_
print "Intercept = ", regr.intercept_
```

We can transform the hyperplane in 3 dimensions into the height by weight 2 dimensional representation by finding where the hyperplane intersects our height by weight plane.  The slopes and intercept in 3 dimensions have a relationship to where they intersect our height by weight plane, which we can get by doing the following (don't worry if you don't understand this transformation at the moment).

```python
slope = - regr.coef_[0,0] / regr.coef_[0,1]
intercept = - regr.coef_[0,1] / regr.intercept_[0]
print slope, intercept

separator_x = np.linspace(height.min(), height.max(), 100)
separator_y = slope * separator_x + intercept
```

And lets see what the separating line looks like.

```python
plt.figure(figsize=(12,12))
plt.plot(height[male_indxs], weight[male_indxs], '.', label='Male', alpha=0.15)
plt.plot(height[female_indxs], weight[female_indxs], '.', label='Female', alpha=0.15)
plt.plot(separator_x, separator_y, 'k-')
plt.legend()
plt.xlabel('Height (in)')
plt.ylabel('Weight (in)')
plt.grid('on')
```

Looks pretty good.  We can easily see what the model will predict for some new, unseen data points, using the fitted model, like this (remember 0.0 means a prediction of Male, and 1.0 means Female).  The line we've drawn has a very fancy-sounding name: the "separating hyperplane."  It's a "separating" hyperplane because it splits the data into two groups: on one side, you guess that someone is female given the height and weight, and on the other you predict male.

```python
#                   in by lbs
new_data = array([ [60., 100.],  # below line, female
                   [60., 150.],  # above line, male
                   [65., 150.],  # etc.
                   [65., 160.],
                   [70., 150.],
                   [70., 175.] ])
p = regr.predict(new_data)
print p
```

Notice that we don't get a 0.0 or 1.0, but real values.  This is because, of course, this is the position of the hyperplane in the z dimension (representing gender) given the x_1, x_2 factors of height and weight.  But since we used 0.0 and 1.0 for our categories, we can simply interpret any prediction < 0.5 as meaning Male, and any > as Female

```python
predictions = (p > 0.5) * 1.0
print predictions
```

We can also use our linear model to see how well it does with the original data in making predictions.  If we subtract the predictions from the correct answers, the result will be 0 if the prediction was correct, and 1 or -1 if incorrect.  We can thus square the differences and
sum them to see how many items were predicted incorrectly from original dataset.

```python
p = regr.predict(factors)
predictions = (p > 0.5) * 1.0
num_errors = np.sum( (gender - predictions)**2.0 )
print "We predicted the following number of data points incorrectly: ", num_errors
ratio_correct = (predictions.shape[0] - num_errors) / predictions.shape[0]
print "This represents a correct classification ratio of: %f (or %0.1f%% correct)" % (ratio_correct, ratio_correct * 100.) 
```

This is a pretty good way to make guesses, for this data set you'd be right about 92% of the time.  This is not bad performance, and for many more complex data classification tasks, we can only dream of achieving this level of accuracy, especially in this case when we are using only 2 simple factors to make our predictions.  In real classification tasks, you'll often have tens, hundreds, or even thousands of potential input factors for predicting classes.  And we often have much more than 2 classes we would like to predict.  This data set just happens to be particularly clean and easy to work with.

You might think that this ratio of correct classification should be related to the $R^2$ score, and it is (recall $R^2$ measure from the previous lecture).  But if you calculated $R^2$ you get the following.  What do you think the difference is between the $R^2$ and the ratio of correct classifications here?

```python
print 'R^2 score of model:', regr.score(factors, gender)
```

General Description of Classification
-------------------------------------

In general, for a  classification task, we have a set of labeled example of the categories we want to learn how to identify.  These examples consist of a label, which we'll also call a class or type, and a series of measured variables that describe each example.  We call these measurements features or predictors.  So previously, height and weight were used as features, and the category we were predicting was the gender with a set of 2 labels: `male` or `female`.  We create a model from the labeled examples that predicts, as best as possible, the label given a set of features.  Because we train the model in a classification task on a set of given labels, this is an example of a supvised learning method.

Classification tasks are simple to understand, but are very important in actual practice to improve the model and get the best predictions possible given the previous data examples.

* Given the results of a mammogram, does a patient have breast cancer or not?
* Given a financial profile of an individual, is the person a high or low risk to default on a loan?
* Given a set of movie ratings, is an individual likely to rate an unseen movie as good, or bad (or 1,2,3 or 4 stars, for an example of a classification with more than 2 categories).


Some examples of classification methods that are commonly used include linear regression (which we just saw an example of) and Logistic Regression (Lecture 5), Nearest-Neighbor methods (Lecture 8), decision trees and support vector machines SVM (both of which I am thinking of covering in second half of course).  Another classification method you might see used frequently are Bayesian Classifiers.

Bayesian Classifier
-------------------

I won't go into the details of how a bayesian classifier works, beyond stating that it relies on probabilities of features observed from past labeled examples, to compute a conditional probability using the famous Bayes formula for conditional probabilities.  Enough detail to roughly understand was is going on can be gleaned from reading the Wikipedia page on the [naive bayes classifier](http://en.wikipedia.org/wiki/Naive_Bayes_classifier).  I will simply use the scikit-learn example of the naive bayes classifier in the following, to see another example of a classifier.

Lets first look at the dataset used in the example.  Scikit-learn provides many interesting and well studied datasets in order to implement examples of their models.

```python
from sklearn import datasets
iris = datasets.load_iris()
print iris.DESCR
```

```python
print type(iris.data), type(iris.target)
print iris.data.shape, iris.target.shape
iris_df = pd.DataFrame(iris.data, columns=['sepal length', 'sepal width', 'petal length', 'petal width'])
iris_df['iris type'] = pd.Series(iris.target, index=iris_df.index)
print iris_df.describe()
print iris_df.head()
```

From the description of the data set, and looking at the object we get, we see that this data set consists of only 150 instances.  There are 3 classes that we are trying to predict, and each class is represented by 50 instances.  For the input or features, there are 4 separate features, described in the description.  These are measured characteristics of an iris (a type of [flower](http://en.wikipedia.org/wiki/Iris_(plant))), such as the petal length and sepal length.  From these 4 features, we can determine which of 3 different types of iris the measure plant is.

I put the data into a pandas dataframe, so that we could confirm that all of the summary statistics are the same as in the description.  I also printed out the first 5 items, just so you could get a feel for the data.  Notice that the iris type is encoded as 0, 1 or 2 for the 3 different classes of iris.

Then using Scikit-learn to build a naive bayes classifier is simple, and of the same pattern as we have seen to build a linear regression model.  We simply create an instance of the model object from the sklearn library, fit it to the data, and use this
to do prediction.  It is able to get 96% correct when tested on the original data.  The classifier built, as mentioned, used the 4 features and created estimates of probability of an iris being of a particular type, given the past probability of the attributes that have been seen for the labeled classes.

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(iris.data, iris.target)
y_pred = gnb.predict(iris.data)
print("Number of mislabeled points : %d" % (iris.target != y_pred).sum())
```

Now lets try some further testing.  Lets see what happens if we pick 5 entries at random from the data, add a bit of noise to them,
and then look at how the prediction does.


```python
random_i = np.random.randint(0, iris.data.shape[0], size=(10,)) # select indexes at random from data
orig_data = iris.data[random_i,:]
noisy_data = orig_data + np.random.standard_normal(size=orig_data.shape) * 0.1 # add noise with mean 0.0 and std of 0.1
orig_target = iris.target[random_i]

print orig_data
print noisy_data
print orig_target

y_pred = gnb.predict(orig_data)
print "Number of mislabeled points originally: %d" % (orig_target != y_pred).sum()
y_pred = gnb.predict(noisy_data)
print "Number of mislabeled points in noisy data: %d" % (orig_target != y_pred).sum()
```

You can try and see what happens if you add more noise to the data to the successful prediction rate.

Linear Classifier using Average Center of Each Category
-------------------------------------------------------

The following is a simplier method for developing a linear separating boundary between two sets of data.  You don't have to follow this, I am still working on polishing it up for future.  If curious, we simply calculate the centroid of the points in each of the labeled class.  We make a decision for new points very simply, depending on which centroid they are closer to.  This can also be described as a line that exactly bisects the line formed between the two centroids.  As you can see, though, the separating hyperplane is inferior to that produced by linear regression (we will get many more misclassifications from this method).

```python
m_height = height[male_indxs].mean()
m_weight = weight[male_indxs].mean()

f_height = height[female_indxs].mean()
f_weight = weight[female_indxs].mean()

print m_height, m_weight
print f_height, f_weight

mid_x = (m_height + f_height) / 2.0
mid_y = (m_weight + f_weight) / 2.0

con_slope = (f_height - m_height) / (f_weight - m_weight)
slope = - 1 / con_slope
intercept = mid_y - slope*mid_x # y = mx + b,  b = y - mx
x = np.linspace(54.0, 79.0, 100)
y = slope * x + intercept
```

```python
plt.figure(figsize=(12,12))
plt.plot(height[male_indxs], weight[male_indxs], '.', label='Male', alpha=0.15)
plt.plot(height[female_indxs], weight[female_indxs], '.', label='Female', alpha=0.15)

plt.plot(m_height, m_weight, 'ko', markersize=10.0, label='Male Centroid')
plt.plot(f_height, f_weight, 'ks', markersize=10.0, label='Female Centroid')

plt.plot(x, y, 'k-')

plt.legend()
plt.xlabel('Height (in)')
plt.ylabel('Weight (in)')
plt.grid()
```
