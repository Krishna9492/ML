```python
def squared_error(inddata, depdata, slope, intercept):
    """inddata is the independata data from which we will make predictions (a list or NumPy array).
    depdata is the data we will be prediting (also a list or NumPy array).
    We calculate a predicited value using column 1 as input, and slope,intercept as parameters of a linear fit.
    The sum squared error is simply the differences of the prediction from the 2 column actual values, squared and summed.

    returns the sum squared error
    """
    predictions = slope + intercept * inddata
    errors = depdata - predictions
    return np.sum(errors**2.0)
```

Lets actually try and evaluate squared_error for both dimensions at the same time.  We will need to make a 3-dimensional plot, showing the intercept and slop on x,y axis, with z being the calculated squared_error for each point on grid.
